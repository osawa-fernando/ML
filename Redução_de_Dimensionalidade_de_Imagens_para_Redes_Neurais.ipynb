```python
#bibliotecas para conversão de tipos de imagens
from PIL import Image
from google.colab import files

```

```python
# RGB para Tons de Cinza
# Upload da imagem .jpg
print("Envie uma imagem .jpg...")
uploaded = files.upload()
nome_arquivo = next(iter(uploaded))

# Abrir imagem e converter para RGB (caso seja RGBA ou L)
img = Image.open(nome_arquivo).convert("RGB")
largura, altura = img.size
pixels = list(img.getdata())

# Função de conversão RGB → tons de cinza
def rgb_para_cinza(r, g, b):
    return int(0.299 * r + 0.587 * g + 0.114 * b)

# Converter todos os pixels
pixels_cinza = []
for r, g, b in pixels:
    gray = rgb_para_cinza(r, g, b)
    pixels_cinza.append((gray, gray, gray))

# Salvar a imagem em formato PPM (modo texto - P3)
nome_saida = "imagem_cinza.ppm"
with open(nome_saida, "w") as f:
    f.write("P3\n")
    f.write(f"{largura} {altura}\n")
    f.write("255\n")
    for r, g, b in pixels_cinza:
        f.write(f"{r} {g} {b}\n")

# Abrir o .ppm gerado e converter para JPG
img_ppm = Image.open(nome_saida)  # nome_saida = "imagem_cinza.ppm"
img_jpg = img_ppm.convert("RGB")
img_jpg.save("husky_cinza.jpg", "JPEG")

# Baixar o .jpg resultante
files.download("husky_cinza.jpg")
```

```python
# Tons de Cinza para P&B
# Abrir imagem cinza .jpg e converter para RGB
img_cinza_jpg = Image.open("husky_cinza.jpg").convert("RGB")
largura, altura = img_cinza_jpg.size
pixels = list(img_cinza_jpg.getdata())

# Converte para imagem preta e branca (limiar)
def cinza_para_pb(r, g, b, limiar=128):
    # Como a imagem já é cinza, R=G=B
    return 255 if r > limiar else 0

pixels_pb = []
for r, g, b in pixels:
    bw = cinza_para_pb(r, g, b)
    pixels_pb.append((bw, bw, bw))

# Salvar novo PPM (preto e branco)
nome_ppm_pb = "husky_pb.ppm"
with open(nome_ppm_pb, "w") as f:
    f.write("P3\n")
    f.write(f"{largura} {altura}\n")
    f.write("255\n")
    for r, g, b in pixels_pb:
        f.write(f"{r} {g} {b}\n")

# Converter o .ppm PB para JPG
img_pb = Image.open(nome_ppm_pb).convert("RGB")
img_pb.save("husky_pb.jpg", "JPEG")

# Download
files.download("husky_pb.jpg")
```
